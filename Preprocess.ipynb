```python
#Binarization
from sklearn.preprocessing import Binarizer
```

```python
x=array[:,1:10]
#dataset[:]=dataset[:].replace(numpy.NaN,0)
dataset=read_csv(filename)
print(dataset.shape)
dataNew=read_csv(filename)
array=dataNew.values
x=array[:,1:10]
```

```python
binarizer=Binarizer(threshold=0.0).fit(x)
binaryx=binarizer.transform(x)
```

```python
numpy.set_printoptions(precision=1)
```

```python
print(binaryx)
```

```python
from matplotlib import pyplot as plt
```

```python
p=binaryx[0:100,1:10]
plt.plot(p)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaleX=StandardScaler().fit(x)
rescaledX=scaleX.transform(x)
numpy.set_printoptions(precision=3)
print(rescaledX)
```

```python
p1=rescaledX[0:50,0:100]
plt.plot(p1)
```

```python
#Normalization
from sklearn.preprocessing import Normalizer
```

```python
scaleNormal=Normalizer().fit(x)
normalscaled=scaleNormal.transform(x)
numpy.set_printoptions(precision=3)
```

```python
print(normalscaled)
```

```python
p3=normalscaled[0:20,0:10]
plt.plot(p3)
```
